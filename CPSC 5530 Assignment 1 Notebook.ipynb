## Visualization 1: Min Temperature in Melbourne 1981-1990

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
#import dataset as df1
df1 = pd.read_csv('daily-min-temperatures.csv')

#print df1 head
df1.head()
```

```python
#check Date column type
print(df1.dtypes)

#convert to date time object
df1['Date'] = pd.to_datetime(df1['Date'])
```

```python
#check that conversion was successful. Note we now have 'Date' as a datetime object
print(df1.dtypes)
```

```python
#create column for year, date, and month in df1 dataframe
df1['year'] = pd.DatetimeIndex(df1['Date']).year
df1['month'] = pd.DatetimeIndex(df1['Date']).month
df1['day'] = pd.DatetimeIndex(df1['Date']).day

#check that columns are correctly added by viewing head and tail of dataframe. 
df1.head()
df1.tail()
```

```python
#preprocessing complete, create boxplot to show monthly statistics
sns.boxplot(data = df1, x = 'month', y = 'Temp').set_title("Monthly Temperature in Melbourne Australia 1981-1990")
```

## Visualization 2: Monthly Air Passengers from 1949-1960

```python
#import dataset as df2
df2 = pd.read_csv('AirPassengers.csv')

#view head of dataframe
df2.head()
```

```python
#check data types of attributes
print(df2.dtypes)

#convert month to datetime object
df2['Month'] = pd.to_datetime(df2['Month'])
```

```python
#check conversion was successful
print(df2.dtypes)
```

```python
#rename Month column as Date 
df2 = df2.rename(columns = {"Month": "Date"})

#create columns in df2 for month and year
df2['year'] = pd.DatetimeIndex(df2['Date']).year
df2['month'] = pd.DatetimeIndex(df2['Date']).month

#check head and tail of data frame to ensure correctness
df2.head()
```

```python
#Plot a heat map using hexagons as markers and color a channel to illustrate the number of airline passengers over time
df2.plot.hexbin(x='year', y='#Passengers', gridsize =20)

#add x,y labels and title
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.title("Heat Map of Number of Airline Passengers by Year 1949-1960")
```

# Visualization 3: Google (GOOG) Stock Prices 2022-2023

```python
#import data set as df3
df3 = pd.read_csv("GOOG.csv")

#print header of dataframe
df3.head()
```

```python
#check object types
print(df3.dtypes)

#convert Date to datetime object
df3['Date'] = pd.to_datetime(df3['Date'])
```

```python
#check conversion was successful
print(df3.dtypes)
```

```python
#sort data frame by date
df3 = df3.sort_values(by='Date')

df3.head()
```

```python
#visualize a line plot showing open, high, low, close
#set figure size
plt.figure(figsize = (10,6))
#plot lines for Open, High, Low, and Close
plt.plot(df3['Date'], df3['Open'], label = "Open")
plt.plot(df3['Date'], df3['High'], label = "High")
plt.plot(df3['Date'], df3['Low'], label = "Low")
plt.plot(df3['Date'], df3['Close'], label = "Close")
#Add legend to differentiate between the lines
plt.legend()
#Add title and x,y labels
plt.title("Google Stock Price 2022-2023")
plt.xlabel("Date")
plt.ylabel("Price")
```
